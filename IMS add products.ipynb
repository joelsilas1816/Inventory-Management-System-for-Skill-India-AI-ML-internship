## to read and display the content of the records.json file 

```python
fp = open("records.json","r")

str_products = fp.read()

fp.close()

items = json.loads(str_products,object_hook=lambda d: {int(k) if k.lstrip('-').isdigit() else k: v for k, v in d.items()})

for i in items:
    print(i," : ",items[i],"\n")
```

## to insert n products into the inventory and store the updated inventory in the records.json file

```python
n = int(input("Enter the number of items to be inserted in the inventory : "))
repeat='y'
i=1

while(True and repeat=='y'):
    flag=0
    indicator=0
        
    while(i<=n):
        flag=0
        indicator=0

        print("\nFor product :",i,"\n")
        id=int(input("\nEnter product id to be inserted : "))

        for key in items:
            if(id == key):
                flag=1
                break

        if(flag==1):

            while(True):
                print("\nA product with id",id,"already exists in the inventory.\nHow would you like to proceed?\n1.Enter another id \n2.Delete the already exisiting id\n3.Quit insertion process")

                ch=int(input("Enter your choice : "))
                if(ch==1):
                    break
                elif(ch==2):
                    for key in items:
                        if(id == key):
                            indicator=1
                            del items[key]
                            break

                    if(indicator):
                        print("\nProduct with id as",id,"has been deleted successfully")
                        revise(items)
                        break
                    else:
                        print("\nProduct not found in the inventory")

                    break
                elif(ch==3):
                    break
                else:
                    print("Please enter a valid choice (1/2) : ")
            if(ch==3):
                repeat='n'
                break
                    
        else:
            name = input("Enter the name of the product : ")          
            date_manufacture= input("Enter the date of manufacture of the product : ")
            date_expiry= input("Enter the expiry date of the product : ")
            place_manufacture= input("Enter the place of manufacture of the product : ")
            types = input("Enter the type of the product : ")

            print("\nEnter the cost price of the product")           
            while(True):
                costPrice = int(input())
                if(costPrice<0):
                    print("\nPlease enter a positive value")
                else:
                    break

            print("\nEnter the marked price of the product")               
            while(True):
                markedPrice = int(input())
                if(markedPrice<0):
                    print("\nPlease enter a positive value")
                else:
                    break

            print("\nEnter the discount(%) on the product")               
            while(True):
                discount= int(input())
                if(discount<0):
                    print("\nPlease enter a positive value")
                else:
                    break


            sellingPrice= markedPrice*((100-discount)//100)


            print("\nEnter the quantity of the product")  
            while(True):
                quantity= int(input())
                if(quantity<0):
                    print("\nPlease enter a positive value")
                else:
                    break

            print("\nEnter the weight(kg) of the product")  
            while(True):
                weight= float(input())
                if(weight<0):
                    print("\nPlease enter a positive value")
                else:
                    break

            items[id] = {"Name" : name , "costPrice" :costPrice , "markedPrice" : markedPrice , "sellingPrice" : sellingPrice , "Quantity" : quantity, "date of manufacture" : date_manufacture, "expiry date" : date_expiry, "place of manufacture" : place_manufacture, "weight(kg)" : weight , "type" : types , "discount(%)" : discount}           

            revise(items)
            i+=1
        




    
```

## to update the values of the attributes of the products

```python
def check(n,val):
    if(n==8 and val=="---"):
        return True
    elif(n==7 and val=="---"):
        print("Manufacturing date cannot be left blank")
        return False
    else:
        if(val=="0"):
            print("Date cannot be negative or zero")
            return False
        elif(int(val[6:10])<=0):
            print("Year cannot be negative or zero")
            return False
        elif(int(val[3:5])<=0 or int(val[3:5])>12):
            print("Month cannot be negative or zero or greater than 12")
            return False
        elif((int(val[3:5])==9 or int(val[3:5])==4 or int(val[3:5])==6 or int(val[3:5])==11) and int(val[:2])==31):
            print("This Month has only 30 days so date cannot be 31")
            return False
        elif(int(val[:2])<=0 or int(val[:2])>31):
            print("Date can neither be zero nor exceed 31")
            return False
    return True
    

n=int(input("\nEnter the number of products whose attributes need to be updated : "))
if(n<=0):
    print(n,"cannot be negative or zero")
else:
    for i in range(0,n):
        print("-- Product",i+1,"--")

        while(True):
            flag=0
            id=int(input("Enter product id : "))

            for key in items:
                if(id == key):
                    flag=1
                    break

            if(flag):                
                while(True):
                    m=int(input("\nEnter the number of attributes whose values need to be updated : "))
                    if(m<=0):
                        print(m,"cannot be negative or zero")
            
                    else:
                        for j in range(0,m):
                            print("\n\n--Attribute",j+1,"--")
                            print("\nSelect attribute whose value needs to be changed:\n1.Name of product\n2.Cost Price\n3.Marked Price\n4.Discount(%)\n5.Selling Price\n6.Place of Manufacture\n7.Date of Manufacture\n8.Expiry Date\n9.type\n10.Weight\n11.Quantity")
                            option = int(input("Enter choice : "))  

                            if(option==1):
                                newName=input("Enter new name of the product : ")
                                items[id]["Name"]=newName

                            elif(option==2):
                                newCostPrice=int(input("Enter new cost price of the product : "))
                                items[id]["costPrice"]=newCostPrice

                            elif(option==3):
                                newMarkedPrice=int(input("Enter new marked price of the product : "))
                                items[id]["markedPrice"]=newMarkedPrice

                            elif(option==4):
                                newDiscount=int(input("Enter new discount(%) for the product : "))
                                items[id]["discount(%)"]=newDiscount

                            elif(option==5):
                                newSellingPrice=int(input("Enter new selling price of the product : "))
                                items[id]["sellingPrice"]=newMarkedPrice

                            elif(option==6):
                                newPlace=input("Enter new place of manufacture of the product : ")
                                items[id]["place of manufacture"]=newPlace

                            elif(option==7):
                                while(True):
                                    newDateOfManufac=input("Enter new date of manufacture (DD-MM-YYYY) of the product : ")

                                    if(check(7,newDateOfManufac)):
                                        items[id]["date of manufacture"]=newDateOfManufac
                                        break
                                    else:
                                        print("\nPlease enter a valid date")

                            elif(option==8):
                                while(True):
                                    newExpDate=input("Enter new date of expiry (DD-MM-YYYY) of the product. If expiry date is not applicable then enter --- : ")

                                    if(check(8,newExpDate)):
                                        items[id]["expiry date"]=newExpDate
                                        break
                                    else:
                                        print("\nPlease enter a valid date")


                            elif(option==9):
                                newType=input("Enter new type of the product : ")
                                items[id]["type"]=newType

                            elif(option==10):
                                newWeight=float(input("Enter new weight(kg) of the product : "))
                                items[id]["weight(kg)"]=newWeight

                            elif(option==11):
                                newQuan=int(input("Enter new quantity of the product : "))
                                items[id]["Quantity"]=newQuan
                                
                        break
                break
            else:
                print("\nProduct not found")

```

```python
def sort_opts(param):
    lst=[];

    if(param=="product_id"):
        for i in items:
            lst.append(i)
    else:
        for i in items:
            lst.append(items[i][param])

    while(True):
        sort_choice = int(input("\nEnter choice : 1.Ascending order 2.Descending order : "))
        if(sort_choice==1):
            lst.sort()
            disp_sorted(param,lst)
            break
        elif(sort_choice==2):
            lst.sort(reverse=True)
            disp_sorted(param,lst)
            break
        else:
            print("\nPlease enter a valid choice : ")
```
