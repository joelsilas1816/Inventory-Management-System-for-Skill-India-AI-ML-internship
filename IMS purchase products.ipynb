## importing the necessary libraries

```python
import json
import datetime
```

## main driver code to avail the services of purchase of products

```python

fp = open("records.json","r")
str_products = fp.read()
deficit=list()
empty=list()

fp.close()

items = json.loads(str_products,object_hook=lambda d: {int(k) if k.lstrip('-').isdigit() else k: v for k, v in d.items()})

# To display daily sales report

def display_Sales_Report():
    
    fobj=open("sales report.txt","a")
    
    fobj.write("*************************************************************************************************************************\n")
    fobj.write("# Daily Sales Report ".center(56))
    fobj.write("\n*************************************************************************************************************************\n")
    fobj.write("\n     Time_of_purchase " +" Customer_Name " +" prod_id " + " Name "  + " costPrice" +" discount(%)" + " sellingPrice"  + " quantity"+ " NetSellingPrice " + " type\n" )

    
    fd = open("sales.json","r")
    str_sales = fd.read()

    js_sales = json.loads(str_sales)
    fd.close()
    

    for i in js_sales:
        mystr= str(str(js_sales[i]["Time_of_purchase"])+"     "+str(js_sales[i]["Customer_Name"]).ljust(7)+" "+str(js_sales[i]["product id"])+"    "+str(js_sales[i]["prod_Name"]).ljust(10)+str(js_sales[i]["costPrice"]).ljust(8)+str(js_sales[i]["discount(%)"]).ljust(11)+" "+str(js_sales[i]["sellingPrice"]).ljust(11)+"  "+str(js_sales[i]["quantity"]).ljust(10)+str(js_sales[i]["NetSellingPrice"]).ljust(13)+str(js_sales[i]["type"])+"\n")
        fobj.write(mystr)
        
    if(deficit):
        fobj.write("\n\nThe products with id "+str(deficit)+" are in deficiency in the inventory. It is recommended to add them in the inventory")
    if(empty):
        fobj.write("\nThe products with id "+str(empty)+" are unavailable in the inventory. It is recommended to add them in the inventory\n\n")

    
        
    fobj.close()
    
    


sales=dict()
e = datetime.datetime.now()

# function to purchase products

def purchase_item():
    exp=False
    flag=0
    
    cart=dict()
    repeat='y'
    a=0

    for i in items:
            if(items[i]["Quantity"]==0):
                empty.append(i)
    
    print("\nEnter customer details : \n")
    customer = input("\nName : ")
    age = int(input("\nAge : "))
    phno = input("\nContact number : ")
    email_id = input("\nEmail id : ")

    while(repeat=='y' or repeat=='Y'):

        id = int(input("\nEnter product id of the product to be purchased : "))
        if(1000>id or id>1029):
            print("\nProduct id not found")
            while(True):
                repeat=input("\nDo you want to purchase any other items ? (y/n) : ")
                if(repeat=='y' or repeat=='Y'):
                    break
                elif(repeat=='n' or repeat=='N'):
                    str_items = json.dumps(items)
                
                    fd = open("records.json","w")
                    fd.write(str_items)
                    fd.close()
                    generate_receipt(repeat,cart,customer,age,phno,email_id)
                    break

                else:
                    print("\nPlease enter a valid choice : ")

        else:
            for key in items:
                if(id == key):
                    flag=1
                    break
            if(flag):
                print("\nProduct with id as",id,"has been found successfully")
                exp=False
                if(items[id]["expiry date"]!= "---"):
                    if(int(items[id]["expiry date"][6:10]) <  (e.year)):
                        print ("Sorry, product expired", (e.year) -  int(items[id]["expiry date"][6:10]),"years ago")
                        exp=True
                    elif(int (items[id]["expiry date"][3:5]) <  (e.month)):
                        print ("Sorry, product expired", (e.month) - int (items[id]["expiry date"][3:5]),"months ago")
                        exp=True    
                    elif( int(items[id]["expiry date"][:2]) <  (e.day)):
                        print ("Sorry, product expired", (e.day) - int (items[id]["expiry date"][:2]),"days ago")
                        exp=True
                    elif(int(items[id]["expiry date"][:2]) ==  (e.day)):
                        print ("Sorry, product expired today")
                        exp=True

                    if(exp):
                        while(True):
                            repeat=input("\nDo you want to purchase any other items ? (y/n) : ")
                            if(repeat=='y' or repeat=='Y'):
                                break
                            elif(repeat=='n' or repeat=='N'):
                                str_items = json.dumps(items)
                            
                                fd = open("records.json","w")
                                fd.write(str_items)
                                fd.close()
                                generate_receipt(repeat,cart,customer,age,phno,email_id)
                                break
                            else:
                                print("\nPlease enter a valid choice : ")
                            
                        
                if(items[id]["expiry date"]== "---" or exp==False):
                    if(items[id]["Quantity"]==0):
                        print("\nSorry these products are unavailable.")
                        while(True):
                            repeat=input("\nDo you want to purchase any other items ? (y/n) : ")
                            if(repeat=='y' or repeat=='Y'):
                                break
                            elif(repeat=='n' or repeat=='N'):
                                str_items = json.dumps(items)

                                fd = open("records.json","w")
                                fd.write(str_items)
                                fd.close()
                                generate_receipt(repeat,cart,customer,age,phno,email_id)
                                break

                            else:
                                print("\nPlease enter a valid choice : ")

                    else:
                        quan = int(input("\nEnter quantity : "))
                        if(quan>items[id]["Quantity"]):

                            print("\nThere are only {} items. Deficit is : {}. Do you wnat to purchase all of these?(y/n) : ".format(items[id]["Quantity"],quan-items[id]["Quantity"]))

                            while(True):
                                choice = input();
                                if(choice=='y' or choice=='Y'):

                                    disc = input("\nDiscount to be applied ?(y/n)");
                                    if(disc=='y' or disc=='Y'):
                                        totalSellingPrice = items[id]["sellingPrice"] * items[id]["Quantity"]
                                        cart[id]= { "Quan" : quan , "totsellPrice" : totalSellingPrice , "discount" : items[id]["discount(%)"]}
                                                                      
                                    else:
                                        totalSellingPrice = items[id]["markedPrice"] * items[id]["Quantity"]
                                        cart[id]= { "Quan" : quan , "totsellPrice" : totalSellingPrice , "discount" : "---"}
                                    
                                    a+=1
                                    sales[a]= {"Time_of_purchase":str(datetime.datetime.now()),"Customer_Name":customer,"product id":id , "prod_Name":items[id]["Name"] , "costPrice":items[id]["costPrice"] ,"discount(%)":cart[id]["discount"], "sellingPrice":items[id]["markedPrice"] , "quantity":items[id]["Quantity"], "NetSellingPrice" : totalSellingPrice, "type" :items[id]["type"]}
                                        
                                    print("{} products with id as {} have been purchased successfully. Thank you for trusting us".format(items[id]["Quantity"],id))

                                    items[id]["Quantity"]=0

                                    if id not in empty:
                                        empty.append(id)

                                    while(True):
                                        repeat=input("\nDo you want to purchase more items ? (y/n) : ")
                                        if(repeat=='y' or repeat=='Y'):
                                            break
                                        elif(repeat=='n' or repeat=='N'):
                                            str_items = json.dumps(items)

                                            fd = open("records.json","w")
                                            fd.write(str_items)
                                            fd.close()
                                            generate_receipt(repeat,cart,customer,age,phno,email_id)
                                            break
                                        else:
                                            print("\nPlease enter a valid choice : ")
                                    break
                                elif(choice=='n' or choice =='N'):
                                    print("\nOkay, Sorry for the inconvenience. Thank you.")
                                    deficit.append(id)

                                    while(True):
                                        repeat=input("\nDo you want to purchase any other items ? (y/n) : ")
                                        if(repeat=='y' or repeat=='Y'):
                                            break
                                        elif(repeat=='n' or repeat=='N'):
                                            str_items = json.dumps(items)

                                            fd = open("records.json","w")
                                            fd.write(str_items)
                                            fd.close()
                                            generate_receipt(repeat,cart,customer,age,phno,email_id)

                                            break
                                        else:
                                            print("\nPlease enter a valid choice : ")
                                    break
                                else:
                                    print("\nPlease enter a valid choice(y/n)")

                        else:

                            disc = input("\nDiscount to be applied ?(y/n)");
                            if(disc=='y' or disc=='Y'):
                                totalSellingPrice = items[id]["sellingPrice"] * quan
                                cart[id]= { "Quan" : quan , "totsellPrice" : totalSellingPrice , "discount" : items[id]["discount(%)"]}
                                        
                            else:
                                totalSellingPrice = items[id]["markedPrice"] * quan
                                cart[id]= { "Quan" : quan , "totsellPrice" : totalSellingPrice , "discount" : "---"}
                                
                            a+=1
                            sales[a]= {"Time_of_purchase":str(datetime.datetime.now()),"Customer_Name":customer,"product id":id , "prod_Name":items[id]["Name"] , "costPrice":items[id]["costPrice"] ,"discount(%)":cart[id]["discount"], "sellingPrice":items[id]["sellingPrice"] , "quantity":quan, "NetSellingPrice" : totalSellingPrice, "type" :items[id]["type"]}
                                   

                            print("{} products with id as {} have been purchased successfully".format(quan,id))

                            items[id]["Quantity"]-=quan

                            while(True):
                                repeat=input("\nDo you want to purchase more items ? (y/n) : ")
                                if(repeat=='y' or repeat=='Y'):
                                    break
                                elif(repeat=='n' or repeat=='N'):
                                    str_items = json.dumps(items)

                                    fd = open("records.json","w")
                                    fd.write(str_items)
                                    fd.close()
                                    generate_receipt(repeat,cart,customer,age,phno,email_id)
                                    break
                                else:
                                    print("\nPlease enter a valid choice : ")


            else:
                print("\nProduct not found in the inventory")
                while(True):
                    repeat=input("\nDo you want to purchase any other items ? (y/n) : ")
                    if(repeat=='y' or repeat=='Y'):
                        break
                    elif(repeat=='n' or repeat=='N'):
                        print("Thank you !!! Please come again !!!")
                        str_items = json.dumps(items)

                        fd = open("records.json","w")
                        fd.write(str_items)
                        fd.close()
                        generate_receipt(repeat,cart,customer,age,phno,email_id)

                        break
                    else:
                        print("\nPlease enter a valid choice : ")

    print("\nThank you !!!")
    
# Function to generate a receipt for the customer if he/she purchases any item    
def generate_receipt(repeat,cart,customer,age,phno,email_id):
    
    if( (repeat=='n' or repeat=='N') and cart):
        print("\nHere is your receipt !!! Thank you for trusting us !!!")

        print("*************************************************************************************************************************")
        print("# Customer details ".center(56))
        print("*************************************************************************************************************************")
        print("Name : ".ljust(10),customer)
        print("Age : ".ljust(10),age)
        print("Contact number : ".ljust(10),phno)
        print("Email id : ".ljust(10),email_id)
        print("*************************************************************************************************************************")
        print("# Billing details ".center(56))
        print("*************************************************************************************************************************")
        print("Product id".ljust(15),"Name".ljust(12),"Marked Price".ljust(15),"Discount(%)".ljust(15),"Sellling Price".ljust(16),"Quantity".ljust(12),"Net Selling Price")
        
        totalMRP=0
        totalDiscount=0
        totalIndivSell=0
        totalQuantity=0
        totalSellPrice=0
        totalMRPSell=0
        
        
        for i in cart:
            indivSell=int(cart[i]["totsellPrice"]//cart[i]["Quan"])
            print(str(i).ljust(15),items[i]["Name"].ljust(12),str(items[i]["markedPrice"]).ljust(15),str(cart[i]["discount"]).ljust(15),str(indivSell).ljust(18),str(cart[i]["Quan"]).ljust(12),str(cart[i]["totsellPrice"]))
            totalMRP += int(items[i]["markedPrice"])
            if(cart[i]["discount"]!="---"):
                totalDiscount += int(cart[i]["discount"])
            totalIndivSell += int(indivSell)
            totalQuantity += int(cart[i]["Quan"])
            totalSellPrice += int(cart[i]["totsellPrice"])
            totalMRPSell += int(items[i]["markedPrice"])*int(cart[i]["Quan"])
            
            
        print("*************************************************************************************************************************")
        print("Total".ljust(15),"    ".ljust(12),str(totalMRP).ljust(15),str(totalDiscount).ljust(15),str(totalIndivSell).ljust(18),str(totalQuantity).ljust(12),str(totalSellPrice))
        print("Congrats, you saved Rs.",totalMRPSell-totalSellPrice,"Please come again !!!")
        print("*************************************************************************************************************************")
        
    fp = open("sales.json","w")
    str_sales = json.dumps(sales)

    fp.write(str_sales)

    fp.close()

    fp=open("records.json","w")
    str_items = json.dumps(items)
    fp.write(str_items)
    fp.close()

    display_Sales_Report()

purchase_item()
    
```
